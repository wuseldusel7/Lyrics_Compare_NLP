```python
import requests
import pandas as pd
import numpy as np
import re
from bs4 import BeautifulSoup
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
```

```python
url = 'https://www.lyrics.com/artist/Coldplay/435023'
```

```python
response = requests.get(url, headers=headers)
```

```python
cold_soup = BeautifulSoup(response.text)
```

```python
def pp(soup):
    print(soup.prettify())
```

```python
pp(cold_soup)
```

```python
from urllib.request import Request, urlopen

req = Request('https://www.lyrics.com/artist/Coldplay/435023')
html_page = urlopen(req)

soup = BeautifulSoup(html_page, "lxml")

links = []
for link in soup.findAll('a'):
    links.append(link.get('href'))

print(links)
```

```python
links[59]
```

```python
re.findall(r'^/lyric', links[59])[0]
```

```python
lst = []
for link in links:
    if re.findall(r'^/lyric', link):
        lst.append(link)
    else:
        continue
```
