```python
import requests
import pandas as pd
import numpy as np
import re
```

```python
url = 'https://www.azlyrics.com/c/coldplay.html'
```

```python
response = requests.get(url)
```

```python
print(response.text)
```

```python
f = open('Coldplay.txt', 'w')
f.write(response.text)
f.close()
```

```python
links = re.findall('href.{100}', response.text)
```

```python
from bs4 import BeautifulSoup
import urllib.request

links = []

html_page = urllib.request.urlopen(url)
soup = BeautifulSoup(html_page, "html.parser")
for link in soup.find_all('a'):
    if '/coldplay/' in str(link.get('href')):
        links.append(link.get('href'))
    else:
        continue
```

```python
for link in soup.find_all('div', class_="listalbum-item" ):
    print(link.get('href'))
```

```python
links
```

```python
for counter, link in enumerate(links):
    url = url + link
    response = requests.get(url)
    open(f'{counter}.txt', 'w').write(response.text)
    print(counter, link, response.status_code)
```
